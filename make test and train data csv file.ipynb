## This module is for preparing the csv file to be fed into a program which will convert the csv file into a tfrecord file(which is used to test the model)

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
annotation_path='C:/Users/sinayak/Project/models-master/research/object_detection/images/labels.csv'
# read the file

df = pd.read_csv(annotation_path, sep = " ", header = None, names = ["frame", "xmin", "ymin", "xmax", "ymax", "occluded", "label","attributes"])
df.head()
```

```python
# add image size column
# as all images are of the same size 1920x1200 we are using only one width and height
# can also do cv2.imread() and then np.shape to get size of image
width=1920
height=1200
df['width']=pd.Series(width*np.ones(len(df)).astype('int'),index=df.index)
df['height']=pd.Series(height*np.ones(len(df)).astype('int'),index=df.index)
df.head()
```

```python
# drop attributes and occluded
df=df.drop(['occluded', 'attributes'], 1)
df.head()
```

```python
# rearrange the columns
df=df[['frame','width','height','label','xmin','ymin','xmax','ymax']]
df.head()
```

```python
# rename the columns names
df.columns=['filename','width','height','class','xmin','ymin','xmax','ymax']
df.head()
```

```python
# remove biker and traffice light labels as they are not in the train dataset
b=df[df['class']=='biker'].index
t=df[df['class']=='trafficLight'].index
df=df.drop(b[i] for i in range(len(b)))
df=df.drop(t[i] for i in range(len(t)))
df.groupby(['class']).size()
```

### The final csv file format

```python
df.to_csv('C:/Users/sinayak/Downloads/labels_new.csv',encoding='utf-8',index=False)
```
